```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generating a normally distributed data of size = 10000, mean=50 and standard dev=100
sample = np.random.normal(50, 100, size=1000)
plt.hist(sample, bins=20)
```

```python

from scipy.stats import norm
#Calculating mean and std of the original data
sample_mean = sample.mean()
sample_std = sample.std()
#Generating 100 uniform points within the range of our data. These points will be the x coordinates
value = np.linspace(sample.min(), sample.max(), 100)
#Calculating the probability densities of these 100 values 
dist = norm.pdf(value, loc=sample_mean, scale=sample_std)
plt.plot(value, dist)
```

```python
#KDE:
#Generating an 100 X 1 matrix of random values bewteen 0 and 1
data = np.random.rand(100, 1)
arr = data.flatten()
plt.hist(arr, bins=20)
```

```python
#Training the gaussian model
from sklearn.neighbors import KernelDensity
model = KernelDensity(kernel='gaussian', bandwidth=0.02)
model.fit(data)                 #Note: The data must always be 2D
```

```python
#Generating 70 uniform values within the range of the original data . These values will be x coordinate values
x = np.linspace(arr.min(), arr.max(), 70)
x = x.reshape(70, 1)
#The y value of the Kernal Density distribution
y = model.score_samples(x)
plt.plot(x, y)
```
