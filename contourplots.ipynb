```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#Loading data
df = pd.read_excel("iris-train.xlsx")
df.head()
```

```python
#For understanding how much each feature contributes to the ml algorithm we can plot a kde
sns.kdeplot(data=df, x='sepallength', hue='class')
```

```python
#Sepalwidth plot
sns.kdeplot(data=df, x='sepalwidth', hue='class')
```

```python
#petallength plot
#Best feature for the ml algorithm 
sns.kdeplot(data=df, x='petallength', hue='class')
```

```python
#petal width
sns.kdeplot(data=df, x='petalwidth', hue='class')
```

```python
#2D plots to understand how 2 features contribute in ml algorithms
#The color represents the 3rd dimesnion, more dense the color , more is its probability density
```

```python
sns.jointplot(data = df, x= 'petallength', y = 'petalwidth', kind='kde', color='skyblue', fill=True, hue='class')
```
